```python
# ****Import packages - had to run this for yfinance to work: pip install yfinance --upgrade --no-cache-dir
import yfinance as yf
import pandas as pd
import datetime as dt

# We want the past 3 years of data from now [or whenever the current date]
start_date = dt.datetime(2018, 10, 1)
end_date = dt.datetime.now()

# Define the ticker list
tickers_df = pd.read_csv("sandp500.csv")
tickers_list = tickers_df["Ticker"].tolist()

# Create placeholder for data
data = pd.DataFrame(columns=tickers_list)

# Fetch the data
for ticker in tickers_list:
    data[ticker] = yf.download(ticker, 
                               start_date,
                               end_date)['Adj Close']
    
# Print first 5 rows of the data
data.head()
```
